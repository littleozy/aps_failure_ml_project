## Setting up null models (classifying all samples as positive/negative) 

Since null models can be done without any modelling, ideally , our machine learning model must have a lower test cost compared to null models (else our models provide no values).


```python
positive_labels = {'positive_predictions':['pos']*len(test)}
positive_predictions = pd.DataFrame(positive_labels)

negative_labels = {'negative_predictions':['neg']*len(test)}
negative_predictions = pd.DataFrame(negative_labels)
```

## Calculate Cost when predicting all variables to be positive

we get total misclassification cost to be  $156250

```python
cost(test['class'],positive_predictions)
```

## Calculate Cost when predicting all variables to be negative

we get total misclassification cost to be  $187500

```python
cost(test['class'],negative_predictions)
```

# Establishing a kill metrics

So here, our kill metrics will be the minimum between the null model misclassification cost = min(187500, 156250) = $156250. And also note by comparing the null models, we see that its better to assume all samples here are positives and sending them to the mechanics , safety first y'all!

Meaning that we will reject the model, given that when the model produces a test cost metrics > $156250.
